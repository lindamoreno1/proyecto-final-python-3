```python
df=pd.read_csv("datos-clima.csv", sep=",", header=None, names=["Fecha","Grados Celsius","Grados Fahrenheit","Condicion"])
dfcopia=df
```

```python
for n in dfcopia["Fecha"]:
    item=n[-18:-7]
    dfcopia["Fecha"]=dfcopia["Fecha"].replace(n,item)

for x in dfcopia["Grados Fahrenheit"]:
    g=x[-4:-2]
    grado=int(g)
    if grado<10: grado=grado+100
    dfcopia["Grados Fahrenheit"]=dfcopia["Grados Fahrenheit"].replace(x,grado)
```

```python
for x in dfcopia["Grados Celsius"]:
    g=x[-4:-2]
    grado=int(g)
    dfcopia["Grados Celsius"]=dfcopia["Grados Celsius"].replace(x,grado)
```

```python
for c in dfcopia["Condicion"]:
    item=c
    item=item.strip("'Condicion': ['")
    item=item[:-4]
    dfcopia["Condicion"]=dfcopia["Condicion"].replace(c,item)
```

```python
dfcopia=dfcopia.reindex(columns=["Fecha","Dia","Hora","Grados Celsius","Grados Fahrenheit","Condicion"])
dfcopia
```

```python
dfcopia.to_csv("clima.csv", index=False)
```
