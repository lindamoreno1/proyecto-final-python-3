# GRAFICOS 

```python
import matplotlib.pyplot as plt
import pandas as pd

df=pd.read_csv("../datos/clima.csv")
```

```python
x=df["Grados Celsius"].unique()
y=df["Grados Celsius"].value_counts().to_list()
plt.bar(x,y)
plt.title("Cuantos dias se registraron cada temperatura")
plt.xlabel("Grados Celsius")
plt.ylabel("Ocurrencias")
plt.show()
```

```python
xmo=df[(df["Dia"]=="Monday")]["Condicion"].unique()
ymo=df[(df["Dia"]=="Monday")]["Condicion"].value_counts().tolist()

xtu=df[(df["Dia"]=="Tuesday")]["Condicion"].unique()
ytu=df[(df["Dia"]=="Tuesday")]["Condicion"].value_counts().tolist()

xw=df[(df["Dia"]=="Wednesday")]["Condicion"].unique()
yw=df[(df["Dia"]=="Wednesday")]["Condicion"].value_counts().tolist()

xth=df[(df["Dia"]=="Thursday")]["Condicion"].unique()
yth=df[(df["Dia"]=="Thursday")]["Condicion"].value_counts().tolist()

xf=df[(df["Dia"]=="Friday")]["Condicion"].unique()
yf=df[(df["Dia"]=="Friday")]["Condicion"].value_counts().tolist()

plt.plot(xmo, ymo, color='red', linestyle='--', marker='o')
for i, txt in enumerate(ymo):
    plt.annotate(txt, (xmo[i], ymo[i]))


plt.plot(xtu, ytu, color='blue', linestyle='--', marker='o')
for i, txt in enumerate(ytu):
    plt.annotate(txt, (xtu[i], ytu[i]))

plt.plot(xw, yw, color='black', linestyle='--', marker='o')
for i, txt in enumerate(yw):
    plt.annotate(txt, (xw[i], yw[i]))

plt.plot(xth, yth, color='green', linestyle='--', marker='o')
for i, txt in enumerate(yth):
    plt.annotate(txt, (xth[i], yth[i]))

plt.plot(xf, yf, color='yellow', linestyle='--', marker='o')
for i, txt in enumerate(yf):
    plt.annotate(txt, (xf[i], yf[i]))

plt.legend(['Lunes', 'Martes','Miercoles','Jueves','Viernes'])


plt.suptitle("CONDICIONES POR DIA")

plt.show()
```

```python
xh = df["Hora"]
yg = df["Grados Celsius"]
colors=[]
for item in yg:
    
    if item ==40: x="DarkRed"
    elif item ==38 or item ==39: x="Red"
    elif item ==36 or item ==37: x="Red"
    elif item ==34 or item ==35: x="Orange"
    elif item ==33 or item ==32: x="Orange"
    elif item ==31 or item ==30: x="Orange"
    elif item ==29 or item ==28: x="Gold"
    elif item ==27 or item ==26: x="Yellow"
    elif item ==25 or item ==24: x="Yellow"
    elif item ==23 or item ==22: x="Yellow"
    colors.append(x)
plt.figure(figsize=(20,10))
plt.scatter(xh, yg,c=colors)

plt.plot()
plt.show()
```
